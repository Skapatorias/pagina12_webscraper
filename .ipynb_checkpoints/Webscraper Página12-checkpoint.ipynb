```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(p12.text, 'lxml')
```

```python
print(s.prettify())
```

```python
 s.find('ul', attrs={'class' : 'horizontal-list main-sections hide-on-dropdown'})
```

```python
 secciones = s.find('ul', attrs={'class' : 'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
secciones
```

```python
seccion = secciones[0]
seccion.find('a')
```

```python
seccion.a.get('href')
```

```python
seccion.a.get_text()
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
```

```python
links_secciones
```

```python
sec = requests.get(links_secciones[0])
```

```python
s_seccion = BeautifulSoup(sec.text, 'lxml')
```

```python
print(s_seccion.prettify())
```

```python
article_list = s_seccion.find('div', attrs={'class': 'articles-list'})
```

```python
lista_notas = []
def obtener():
    for article in article_list.find_all('article'):
        if article.a:
            lista_notas.append(article.a.get('href'))
    return lista_notas
```

```python
lista = obtener()
print (lista)
```

```python
url_nota = lista[0]
url_nota
```

```python
try: 
    nota = requests.get(url_nota)
    if nota.status_code == 200:
        s_nota = BeautifulSoup(nota.text, 'lxml')  
    #Extraemos el titulo
        titulo = s_nota.find('h1', attrs = {'class': 'article-title'})
        print(titulo.text)
            
    #Extraer Fecha
        fecha = s_nota.find('span', attrs= {'pubdate': 'pubdate'}).get('datetime')
        print(fecha)
    
    #Extrer copete
        copete = s_nota.find('div', attrs={'class': 'article-summary'}).text
        print(copete)
    
    #Extraer volanta
        volanta = s_nota.find('h2', attrs={'class': 'article-prefix'}).text
        print(volanta)

        
except Exception as e:
    print('Error')
    print(e)
    print ('\n')
    
```

```python
media = s_nota.find('div', attrs={'class': 'article-main-media-image'})
```
